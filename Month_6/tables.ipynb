```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot_bar_chart(csv_file, columns_to_plot, x_label, high=True):
    df = pd.read_csv(csv_file)
    for col in columns_to_plot:
        if col not in df.columns:
            raise KeyError(f"Column '{col}' not found in the CSV file.")
    values = df[columns_to_plot[0]]
    idx = np.argmin(values)
    bar = plt.bar(x_label, values, width=0.5)
    bar[idx].set_color('red')
    plt.text(
    idx,
    values[idx] + 0.005,  # Adjust y-offset as needed
    f"{values[idx] :.2f}s",  # Format accuracy value
    ha="center",
    va="bottom",
    )
    plt.title(columns_to_plot[0])
    plt.tight_layout()
    plt.show()

```

```python
model_names = ["LSTM 1", "LSTM 2", "GRU 1", "GRU 2", "BiRNN 1", "BiRNN 2", "CRNN 1", "CRNN 2"]
plot_bar_chart('Evaluation_Metrics_1/Metrics_Table', ['Training Accuracy'], model_names, True)
```

```python
plot_bar_chart('Evaluation_Metrics_1/Metrics_Table', ['Validation Accuracy'], model_names, True)
```

```python
plot_bar_chart('Evaluation_Metrics_1/Metrics_Table', ['Testing Accuracy'], model_names, True)
```

```python
plot_bar_chart('Evaluation_Metrics_1/Latency_Table', ['Training Latency'], model_names, True)
```
